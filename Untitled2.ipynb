```python
from google.colab.patches import cv2_imshow
import cv2

def prewitt_edge_detection(image_path):

    img1=cv2.imread(image_path)
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        print("Error: Image not found.")
        return


    Gx = [[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]
    Gy = [[-1, -1, -1], [0, 0, 0], [1, 1, 1]]


    edge_image = image.copy()


    rows, cols = image.shape
    for i in range(1, rows - 1):
        for j in range(1, cols - 1):

            region = image[i-1:i+2, j-1:j+2]


            gx = sum(sum(Gx * region))
            gy = sum(sum(Gy * region))


            magnitude = min(255, abs(gx) + abs(gy))


            edge_image[i, j] = magnitude


    cv2_imshow(img1)
    cv2_imshow(image)
    cv2_imshow(edge_image)

if __name__ == '__main__':
    image_path = '/content/my image.jpeg'
    prewitt_edge_detection(image_path)
```
